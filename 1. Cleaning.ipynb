```python
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
df_test['clean_msg']= df_test['Text'].apply(lambda x:remove_punctuation(x))
df_test.head()
```

```python
df_test['msg_lower']= df_test['clean_msg'].apply(lambda x: x.lower())
```

```python
df_test[['clean_msg','msg_lower']]
```

```python
def tokenization(text):
    tokens = re.split(r"\s+",text)
    return tokens
df_test['msg_tokenied']= df_test['msg_lower'].apply(lambda x: tokenization(x))
```

```python
df_test[['msg_lower','msg_tokenied']]
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
def remove_stopwords(text):
    output= [i for i in text if i not in stop_words]
    return output
```

```python
df_test['no_stopwords']= df_test['msg_tokenied'].apply(lambda x:remove_stopwords(x))
```

```python
df_test[['msg_tokenied','no_stopwords']]
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text
```

```python
df_test['msg_lemmatized']=df_test['no_stopwords'].apply(lambda x:lemmatizer(x))
```

```python
df_test[['msg_tokenied','msg_lemmatized']]
```

```python
df_test["final_rev"] = df_test['msg_lemmatized'].apply(' '.join)
```
